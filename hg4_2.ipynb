```python
import gdown

gdown.download('https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)
```

```python
import pandas as pd

ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(ns_book7['도서권수'],ns_book7['대출건수'])
plt.show()

#Adjusting transparency
#Overlap areas are displayed with darker shading
plt.scatter(ns_book7['도서권수'],ns_book7['대출건수'],alpha=0.1)
plt.show()
```

```python
ns_book7['대출건수'].describe()
```

```python
plt.hist(ns_book7['대출건수'])
plt.show()

#Adjusting intervals : log_scale
plt.hist(ns_book7['대출건수'])
plt.yscale('log')
plt.show()

#Dividing the intervals into finer segments
plt.hist(ns_book7['대출건수'], bins=100)
plt.yscale('log')
plt.show()
```

> **"IQR"** stands for "Interquartile Range." It is a statistical measure used to describe the spread or dispersion of data in a dataset. The IQR is the range between the first quartile (Q1) and the third quartile (Q3) and represents the middle 50% of the data. It is calculated as: IQR = Q3 - Q1







```python
plt.boxplot(ns_book7[['대출건수','도서권수']])
plt.yscale('log')
plt.show() #len line = IQR*1.5

#vert
plt.boxplot(ns_book7[['대출건수','도서권수']],vert=False)
plt.xscale('log')
plt.show()
```

> **"EDA"** stands for "Exploratory Data Analysis." It is an approach to analyzing data sets to summarize their main characteristics, often with the help of data visualization methods. EDA is used to understand the data, discover patterns, relationships, anomalies, and to generate hypotheses that can guide further analysis or decision-making.

```python
import gdown

gdown.download('https://bit.ly/3736JW1', 'ns_book6.csv', quiet=False)
```

```python
import pandas as pd

ns_book6 = pd.read_csv('ns_book6.csv', low_memory=False)
ns_book6.head()
```

```python
ns_book6.describe()
```

```python
ns_book7 = ns_book6[ns_book6['도서권수']>0]
ns_book7.describe()
```

```python
#mean
print(ns_book7['대출건수'].mean())
#median
print(ns_book7['대출건수'].median())
#min
print(ns_book7['대출건수'].min())
#max
print(ns_book7['대출건수'].max())
#quantile
print()
print(ns_book7['대출건수'].quantile([0.25,0.5,0.75]))
```

```python
#Calculate the percentile of a specific number
a_specific_number = 10
borrow_10_flag = ns_book7['대출건수'] < a_specific_number
print(borrow_10_flag.mean())
print(ns_book7['대출건수'].quantile(borrow_10_flag.mean()))
```

```python
#var
print(ns_book7['대출건수'].var())
#std = standard deviation
print(ns_book6['대출건수'].std())
#mode
print(ns_book7['발행년도'].mode())
```

```python
#To calculate descriptive statistics in a pandas dataframe.
ns_book7.mean(numeric_only=True)
```

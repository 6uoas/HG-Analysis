keyword = API, HTTP, JSON, XML

> **API**는 애플리케이션 소프트웨어를 빌드하고 통합하기 위한 정의 및 프로토콜 세트인 애플리케이션 프로그래밍 인터페이스(Application Programming Interface)를 뜻합니다. *RedHat*

> **HTTP**(Hypertext Transfer Protocol)는 클라이언트와 서버 간 통신을 위한 통신 규칙 세트 또는 프로토콜입니다. 사용자가 웹 사이트를 방문하면 사용자 브라우저가 웹 서버에 HTTP 요청을 전송하고 웹 서버는 HTTP 응답으로 응답합니다. *AWS*

웹 기반의 API는 html보단 csv,json,xml을 선호함

> **JSON**(JavaScript Object Notation)은 Javascript 객체 문법으로 구조화된 데이터를 표현하기 위한 문자 기반의 표준 포맷 *Mozilla*

```python
#JSON형식으로 만든 파이썬 딕셔너리
d = {'name':'혼자 공부하는 데이터 분석'}
d['name']
```

```python
#dict d 를 json에 맞게 변화하는 법
import json
d_str = json.dumps(d, ensure_ascii=False) #아스키 문자 외에 다른 문자를 16진수로 출력해서 한글 출력하려면 False
d_str, type(d_str) #json의 타입은 str
```

```python
#json문자열을 파이썬 객체로
d2 = json.loads(d_str)
d2['name'], type(d2)
#프로그램상의 객체를 저장하거나 읽을 수 있는 형태로 변환하는 것 == 직렬화(serialization)
#그 반대 == 역직렬화(deserialization) -> 웹 기반 API에서 전달되는 데이터는 json임 --> 파이썬에서 쓰려면 역직렬화 필요
```

```python
d3_str = """
[
  {"name":"혼자 공부하는 데이터 분석","author":"박해선","year":2022},
  {"name":"혼자 공부하는 머신러닝+딥러닝","author":"박해선","year":2020}
]
"""
#세겹따옴표로 문자열 줄 나눠 표현 가능
d3 = json.loads(d3_str)
d3[1]['name']
```

```python
#json 문자열을 데이터프레임으로
import pandas as pd
pd.read_json(d3_str)
```

```python
#파이썬 리스트->딕셔너리를 데이터프레임으로
pd.DataFrame(d3)
```

> **XML**(Extensible Markup Language)은 데이터를 정의하는 규칙을 제공하는 마크업 언어 *AWS*

```python
#XML문자열을 파이썬 객체로
xml_test = """
<book>
  <name>hg-da</name>
  <author>phs</author>
  <year>2022</year>
</book>
"""

import xml.etree.ElementTree as et
book = et.fromstring(xml_test)
type(book), book.tag #가장 먼저 등장하는 element=tag
```

```python
#자식 엘리먼트

#방법1
print(list(book))
name,author,year = list(book)
print(name.text) #텍스트 출력

#방법2
name = book.findtext('name') #findtext는 해당 엘리먼트 찾아 값 출력
print(name)
```

```python
x2_str = """
<books>
    <book>
        <name>혼자 공부하는 데이터 분석</name>
        <author>박해선</author>
        <year>2022</year>
    </book>
    <book>
        <name>혼자 공부하는 머신러닝+딥러닝</name>
        <author>박해선</author>
        <year>2020</year>
    </book>
</books>
"""
books = et.fromstring(x2_str)
books.tag
```

```python
for book in books:
  print(book)
  print(book.findtext('name'))
```
